```python
import pandas as pd
import numpy as np
data = pd.read_csv("../Originals/Fragmented/PRJNA483794.csv")
data.loc[data["LAT LON"]=='4.023550, 9.209440', "LAT LON"] = '-15.8083697,35.0530646'
for col in data.columns:
    data.loc[pd.isna(data[col]), col] = np.nan
    data.loc[data[col]=="-", col] = np.nan
    try:
        data[col] = data[col].apply(lambda x: x.upper().strip())
        data.loc[data[col]=="", col] = np.nan
    except:
        pass
    
    print(col)
    print("-------------")
    print(set(data[col]), len(set(data[col])))
    
    print("=========================================================")

    
data.to_csv("../Corrected/PRJNA483794.csv")
```
