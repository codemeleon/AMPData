```python
import pandas as pd
import numpy as np
data = pd.read_csv("../Originals/Fragmented/PRJNA285083.csv")
for col in data.columns:
    data.loc[pd.isna(data[col]), col] = np.nan
    data.loc[data[col]=="-", col] = np.nan
    try:
        data[col] = data[col].apply(lambda x: x.upper().strip())
        data.loc[data[col]=="", col] = np.nan
    except:
        pass
    
    print(col)
    print("-------------")
    print(set(data[col]), len(set(data[col])))
    
    print("=========================================================")

    
data.to_csv("../Corrected/PRJNA285083.csv")
```
